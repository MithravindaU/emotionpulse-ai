```python
import streamlit as st
```

```python
happy_words = [
    "happy",
    "excited",
    "great",
    "awesome",
    "good"
]

sad_words = [
    "sad",
    "tired",
    "upset",
    "bad",
    "cry"
]

angry_words = [
    "angry",
    "mad",
    "furious"
]
```

```python
st.title("🎭 AI Mood Detector")
```

```python
user_input = st.text_input(
    "How are you feeling today?"
)
```

```python
if user_input:

    user_input = user_input.lower()

    detected = False
```

```python
    if user_input:
      user_input = user_input.lower()

    detected = False
    for word in happy_words:

        if word in user_input:

            st.success(
                "You seem happy today 😄⚡"
            )

            detected = True

            break
        elif not detected:
          for word in sad_words:

            if word in user_input:

                st.warning(
                    "You seem tired or sad today 😭"
                )

                detected = True

                break
```

```python
    if not detected:

        for word in angry_words:

            if word in user_input:

                st.error(
                    "You seem angry today 😤"
                )

                detected = True

                break
```

```python
    if not detected:

        st.info(
            "Hmm... I can't detect your mood yet 👀"
        )
```
